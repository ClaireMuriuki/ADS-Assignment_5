## Questions

### Show the age distribution using a histogram

```python
hist_tit = px.histogram(df, x = 'Age', nbins = 20, title = 'Titatinic Age Distribution')
hist_tit.show()
```

### Show the age distribution on gender using the histogram

```python
female_hist = px.histogram(df, x = 'Age', nbins=30, color ='Sex', title = 'Titanic Age Distribution on Gender' )
female_hist.show()
```

```python
corr = df.corr()

trace = go.Heatmap(z=corr.values,
                  x=corr.index.values,
                  y=corr.columns.values)
data=[trace]
py.iplot(data, filename='basic-heatmap')
```

### Pivot data setting column Pclass  as the columns and Fare as the values. From the resulting straucture use box plot to show the distribution of its values in 3 columns 

```python
# Select Pclass and Fare
df2 = df[['Pclass', 'Fare']]
# Pivot df2
df2.pivot( columns='Pclass', values='Fare')
```

```python
#Plot boxplot
fig = px.box(df2, x = 'Pclass', y = 'Fare')
fig.show()
```

### Plot value counts of passengers who survived and did not survive based on gender

```python
df3 = df[['PassengerId','Survived', 'Sex', 'Pclass']]
df3
```

```python
df4 = df3.groupby(['Sex'])['Survived'].value_counts()
df4
```
